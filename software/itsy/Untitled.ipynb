```python
import matplotlib.pyplot as plt
import numpy as np
import nco

lut_width = 10
word_width = 16

n = nco.NCO(lut_width, word_width)

vals = np.zeros(2**12)
for i in range(2**12-1):
    vals[i] = n.lookup(i)

bigger = np.concat([vals for x in range(16)])

plt.plot(bigger)
plt.show()
plt.magnitude_spectrum(bigger, scale="dB")
plt.show()
```
